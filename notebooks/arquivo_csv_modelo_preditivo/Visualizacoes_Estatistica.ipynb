```python
import pandas as pd
```

```python
db_inicial = '/content/bancoDeDados.xlsx'
sheet_name = 'produtosLimpa'
df_inicial = pd.read_excel(db_inicial, sheet_name=sheet_name)
df_inicial.tail()
```

```python
df_inicial = df_inicial.loc[:, ~df_inicial.columns.str.startswith('Unnamed')]
df_inicial.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_inicial.count()
```

```python
##aqui podemos observar que existem fornecedores que entregam muito e fornecedores que só entregaram uma fez
school_counts = df_inicial['cie'].value_counts()
print(school_counts)
```

```python
#Aqui podemos observar que existem fornecedores que entregam muito e fornecedors que só entregaram uma vez
fornecedores_count = df_inicial['Fornecedor'].value_counts()
print(fornecedores_count)
```

```python
df_cie = df_inicial['cie'].value_counts()
escolas_maiores = df_cie.head(10)
```

Visualização dos dados:

```python
plt.figure(figsize=(12, 6))  # Ajuste o tamanho da figura
sns.barplot(x=escolas_maiores.index, y=escolas_maiores.values)
plt.xlabel('CIE da escola')
plt.ylabel('Avaliações')
plt.title('Top 10 Escolas que mais avaliam')
plt.xticks(rotation=45)  # Rotacione os rótulos do eixo x para facilitar a leitura
plt.show()
```

```python
df_fornecedores = df_inicial['Fornecedor'].value_counts()
maiores_fornecedores = df_fornecedores.head(10)
```

```python
plt.figure(figsize=(12, 6))  # Ajuste o tamanho da figura
sns.barplot(x=maiores_fornecedores.index, y=maiores_fornecedores.values)
plt.xlabel('Nome do fornecedor')
plt.ylabel('Avaliações')
plt.title('Top 10 Fornecedores que mais entregam')
plt.xticks(rotation=90)  # Rotacione os rótulos do eixo x para facilitar a leitura
plt.show()
```

Estatística descritiva das colunas:

```python
entregas_por_fornecedor = df_inicial.groupby('Fornecedor')['Fornecedor'].count()


total_entregas_por_fornecedor = entregas_por_fornecedor.sum()


num_fornecedores = len(entregas_por_fornecedor)


media_entregas = total_entregas_por_fornecedor / num_fornecedores

print(f"O número total de entregas realizadas por todos os fornecedores é: {total_entregas_por_fornecedor} entregas")
print(f"A média do número de entregas por fornecedor é: {media_entregas} entregas")
```

```python
CIEs = df_inicial.groupby('cie')['cie'].count()



numero_de_avaliacoes = CIEs.sum()


num_num_de_aval = len(CIEs)


media_entregas = numero_de_avaliacoes / num_num_de_aval

print(f"O número total de avaliações realizadas por todas as escolas é: {numero_de_avaliacoes} avaliações")
print(f"A média do número de avaliações por escola é: {media_entregas} avaliações")
```
