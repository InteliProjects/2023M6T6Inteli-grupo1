```r
from surprise import Dataset, Reader

reader = Reader(rating_scale=(0, 5))  # Defina a escala de classificação de acordo com seus dados
data = Dataset.load_from_df(user_item_matrix.stack().reset_index(), reader)
```

# Divisão do conjunto entre treino e teste e processo de fitting do modelo:

```r
from surprise.model_selection import train_test_split

trainset, testset = train_test_split(data, test_size=0.2)
```

```r
from surprise import SVD  # Substitua pelo algoritmo de sua escolha

model = SVD()  # Crie uma instância do modelo
model.fit(trainset)
```

# Interpretação das métricas obtidas após o treinamento do modelo:

Para avaliar o modelo de recomendação em um conjunto de teste, utilizamos duas métricas de desempenho são calculadas: RMSE (Root Mean Squared Error) e MAE (Mean Absolute Error). O RMSE mede a média das diferenças ao quadrado entre as previsões e os valores reais, dando mais peso a erros maiores. Por outro lado, o MAE mede a média das diferenças absolutas, tratando erros grandes e pequenos igualmente. Ambas as métricas fornecem insights sobre o quão bem o modelo se ajusta aos dados de teste, sendo úteis para avaliação e comparação de modelos de recomendação. Os resultados são impressos para facilitar a interpretação do desempenho do modelo. O código printa duas vezes pois anteriormente o código pode ter sido executado mais uma vez.

```r
from surprise import accuracy

predictions = model.test(testset)

rmse = accuracy.rmse(predictions)  # Calcule a raiz do erro quadrático médio
mae = accuracy.mae(predictions)    # Calcule o erro absoluto médio

print(f'RMSE: {rmse:.2f}')
print(f'MAE: {mae:.2f}')
```

# Extração de Principais Recomendações

A função foi criada para extrair as principais recomendações de um conjunto de previsões. A entrada para a função é uma lista de previsões, onde cada entrada contém informações como o ID do usuário (uid), o ID do item (iid), o valor verdadeiro (true_r) e a estimativa do modelo (est). A função organiza essas previsões em um dicionário chamado top_n, e os valores são listas de tuplas contendo os IDs dos itens e as estimativas correspondentes. Em seguida, para cada usuário, as recomendações são classificadas com base nas estimativas em ordem decrescente, e apenas as principais recomendações são retidas. O resultado final é um dicionário que mapeia IDs de usuários para listas das principais recomendações.

```r
def get_top_n(predictions, n=10):
    top_n = {}
    for uid, iid, true_r, est, _ in predictions:
        if uid not in top_n:
            top_n[uid] = []
        top_n[uid].append((iid, est))

    for uid, user_ratings in top_n.items():
        user_ratings.sort(key=lambda x: x[1], reverse=True)
        top_n[uid] = user_ratings[:n]

    return top_n

top_n_recommendations = get_top_n(predictions)
```

# Criação do dicionário com os melhores fornecedores:

 O código cria um dicionário, media_notas, para calcular e armazenar as médias de notas atribuídas por escolas a fornecedores. Itera sobre o dataframe, preenchendo o dicionário com notas maiores que zero. Em seguida, determina o melhor fornecedor para cada escola com base na média mais alta das notas atribuídas, armazenando essas informações no dicionário melhores_fornecedores. O resultado final é a impressão do dicionário, revelando qual fornecedor tem a nota média mais alta para cada escola. Esse código é útil para analisar e otimizar as relações entre escolas e fornecedores com base nas avaliações atribuídas.

```r
from collections import defaultdict

# Crie um dicionário para armazenar as médias de notas para cada escola e fornecedor
media_notas = defaultdict(dict)

# Itere por todas as entregas no DataFrame (supondo que seu DataFrame seja chamado 'user_item_matrix')
for cie, row in user_item_matrix.iterrows():
    for cnpj, nota in row.items():
        if nota > 0:  # Certifique-se de que a nota seja maior que zero
            if cie not in media_notas:
                media_notas[cie] = {}
            if cnpj not in media_notas[cie]:
                media_notas[cie][cnpj] = []

            media_notas[cie][cnpj].append(nota)

# Encontre o melhor fornecedor para cada escola (o fornecedor com a nota média mais alta)
melhores_fornecedores = {}
for cie, fornecedores in media_notas.items():
    melhor_fornecedor = max(fornecedores, key=lambda c: sum(fornecedores[c]) / len(fornecedores[c]))
    melhores_fornecedores[cie] = melhor_fornecedor

#print(melhores_fornecedores)
```

# Função que utiliza o modelo exportado:

Nessa função o usuário insere o CIE de alguma escola e o modelo encontra o fornecedor com a nota mais alta para esse CIE. Caso ele não encontre o CIE em questão ele retorna que o cie não foi encontrado

```r
# Função para encontrar o melhor fornecedor com base no CIE inserido pelo usuário
def encontrar_melhor_fornecedor(cie_inserido):
    cie_inserido_float = float(cie_inserido)  # Converte a entrada para float
    if cie_inserido_float in melhores_fornecedores:
        melhor_fornecedor = melhores_fornecedores[cie_inserido_float]
        return f"O melhor fornecedor para a escola com CIE {cie_inserido_float} é o CNPJ {melhor_fornecedor}."
    else:
        return f"O CIE {cie_inserido_float} não foi encontrado no conjunto de dados."

# Solicite ao usuário inserir o CIE
cie_inserido = input("Insira o CIE da escola: ")

# Chame a função para encontrar o melhor fornecedor
resultado = encontrar_melhor_fornecedor(cie_inserido)
print(resultado)
```
